```python
#csv Mortos (raw)
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")

#csv Casos recuperados (raw)
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")

display(deaths.head())
display(recovered.head())
```

```python
#funcao que retorna os dados mais recentes
def latest_by_country(data):
  return data.groupby("Country/Region").sum()[last_date].head()
```

```python
#Selecionando a ultima coluna 
def latest_by_country(data):
  return data.groupby("Country/Region").sum().iloc[:,-1]

display(latest_by_country(confirmed).head())
display(latest_by_country(deaths).head())
display(latest_by_country(recovered).head())
```

```python
# Concatenando as ultimas colunas dos tres dados
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
combined = pd.concat(informations, axis=1)
combined.columns = ["confirmed", "deaths", "recovered"]
combined.head()
```

```python
#taxa de letalidade é expressa em %
# o numero de casos letais dentro dos casos foram letais e não letais

# Imagina que existe uma doença que em janeiro de 2020 afetou 100 pessoas
# 5 pessoas morreram (por exemplo). A letalidade foi de 5%
# a taxa não é 5% para sempre porque a doença muda e continua existindo

#conclusão: muito dificil de falar de letalidade tentando ser preciso durante a epidemia.

# no começo é comum que somente casos graves sejam reportados
```

```python
#Somar as colunas
sum_up = combined.sum()
sum_up
```

```python
# Estimativa de letalidade => mortes / confirmados
sum_up = combined.sum()
letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
print(letality_rate_1 * 100)
```

```python
#premissas:
# UN China = John Hopkins China

#Onde tinha mainland china renomeado de china
def rename_location (location):
  if location == 'China':
    return 'China'
  if location == 'Republic of Korea':
    return 'South Korea'
  if location == 'United States of America':
    return 'Iran'
  if location == 'Iran (Islamic Republic of)':
    return 'US'
  if location == 'China Hong Kong SAR':
    return 'Hong Kong'
  return location  

un_population_total['location_for_who'] = un_population_total.index.map(rename_location)
```

É muito ruim e dificil fazer comparcao com periodos de tempos difererntes(2 meses / 12 meses). Você quer extrapolar simplesmente e multiplicar por 6? 24 a mortalidade. Mas extrapolar com esse tipo de crescimento pode ser muito irreal. Entao infelizmente tem que ir por estimativas... estatistica. 

```python
grouped_per_day = pd.DataFrame([confirmed.sum()[3:-2],recovered.sum()[3:],deaths.sum()[3:]],
                                  index=["confirmed", "recovered", "deaths"]).T
grouped_per_day
```

```python
#taxa de letalidade

grouped_per_day["letality_rate_1"] = grouped_per_day["deaths"] / grouped_per_day["confirmed"] * 100
grouped_per_day["letality_rate_2"] = grouped_per_day["deaths"] / (grouped_per_day["deaths"] + grouped_per_day["recovered"]) * 100

grouped_per_day.head()
```

```python
grouped_per_day["letality_rate_1"].plot()
grouped_per_day["letality_rate_2"].plot()
plt.title("Letalidade 'pessimista e otimista' no mundo todo")
plt.show()
```

```python
#melhorar o grafico
plt.figure(figsize=(10,5))
grouped_per_day["letality_rate_1"].plot()
ax = grouped_per_day["letality_rate_2"].plot()
plt.legend(['Mortes / casos confirmados', 'Mortes / (Mortes + Recuperados)'] )
plt.title("Letalidade 'pessimista e otimista' no mundo todo, convergindo para um mesmo valor")
plt.grid(linestyle='-')
ax.set_ylabel("% letalidade")
ax.set_xlabel("Fonte: curso www.alura.com.br e dados da Johns Hopkins: https://github.com/CSSEGISandData/COVID-19 ")
plt.show()
```
