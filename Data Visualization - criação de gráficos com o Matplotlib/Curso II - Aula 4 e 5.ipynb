## Analise de dados de Flores

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('dados/iris.csv')
```

```python
df.head()
```

```python
#criando gráfico de dispersão
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Gráfico de dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da sépala', fontsize=20)
eixo.set_ylabel('Largura da sépala', fontsize=20)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Gráfico de dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)
eixo.tick_params(labelsize=15)
```

```python
#colorir de acordo com a especies usando dicionario
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'red', 'Iris-versicolor': 'blue', 'Iris-virginica': 'green'}

#plotando o gráfico no laço for como filtro para selecionar as flores
for especie in df ['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color = cores[especie])

eixo.set_title('Gráfico de dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)
eixo.tick_params(labelsize=15)
eixo.legend(cores, fontsize=20)
```

```python
#marcadores
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'red', 'Iris-versicolor': 'blue', 'Iris-virginica': 'green'}
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor': 'o', 'Iris-virginica': 'v'}

#add marker e tamanho(s)
for especie in df ['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                color = cores[especie], marker=marcadores[especie], s=100)

eixo.set_title('Gráfico de Dispersão', fontsize=25, pad=15)
eixo.set_xlabel('Comprimento da sépala', fontsize=15)
eixo.set_ylabel('Largura da sépala', fontsize=15)
eixo.tick_params(labelsize=15)
eixo.legend(cores, fontsize=20)
```

```python
#criar box plot da largura da petala
fig = plt.figure(figsize=(6,4))
eixo  = fig.add_axes([0,0,1,1])

eixo.boxplot(df['largura_pétala'])
eixo.set_title('Gráfico de caixa', fontsize=15, pad=10)
eixo.set_xticklabels(['largura_pétala'])
```

```python
fig = plt.figure(figsize=(8,6))
eixo  = fig.add_axes([0,0,1,1])

#sem especie
eixo.boxplot(df.drop('espécie', axis=1).values)
eixo.set_title('Gráfico de caixa', fontsize=15, pad=10)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)
```

```python
#Cores
fig = plt.figure(figsize=(8,6))
eixo  = fig.add_axes([0,0,1,1])

eixo.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)
eixo.set_title('Gráfico de caixa', fontsize=15, pad=10)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)
```

```python
#Cores2
fig = plt.figure(figsize=(8,6))
eixo  = fig.add_axes([0,0,1,1])

cores = ['red','blue', 'green','orange']

caixas = eixo.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)
eixo.set_title('Gráfico de caixa', fontsize=15, pad=10)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)    
```

```python
#trocar marcadores
fig = plt.figure(figsize=(8,6))
eixo  = fig.add_axes([0,0,1,1])

cores = ['red','blue', 'green','orange']

caixas = eixo.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)
eixo.set_title('Gráfico de caixa', fontsize=15, pad=10)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)

for outlier in caixas['fliers']:
    outlier.set(marker='x', markersize=8)
```

```python
#criando histogramas
fig = plt.figure(figsize=(8,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'])
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)
```

```python
#aumentar detalhes de colunas
fig = plt.figure(figsize=(8,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)
```

```python
#inverter histograma
fig = plt.figure(figsize=(8,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins=20, orientation='horizontal')
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)
```

```python
#normalizar histograma (obs y)
fig = plt.figure(figsize=(7,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins=20, density=True)
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)
```

```python
#media (mu) e desvio padrão(sigma)
fig = plt.figure(figsize=(7,4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma),
             xy = (4.5,20), fontsize = 18)
```

```python
#desenhar reta para indicar media e mediana
fig = plt.figure(figsize=(7,4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma),
             xy = (4.5,20), fontsize = 18)

#media
eixo.axvline(mu, color = 'red', linestyle = '--')
eixo.annotate('média', xy=(mu-1.3,28), fontsize= 18)
```

```python
fig = plt.figure(figsize=(7,4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins=20)
eixo.set_title('Histograma', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
             xy = (4.5,20), fontsize = 18)

#media
eixo.axvline(mu, color = 'red', linestyle = '--')
eixo.annotate('média', xy=(mu-1.3,28), fontsize= 18)

#mediana
eixo.axvline(df['comprimento_pétala'].median(), color = 'orange', linestyle='--')
eixo.annotate('mediana', xy=(df['comprimento_pétala'].median(), 31), fontsize=18)
```

```python
#comparação de todas as especies em uma única imagem

#para Iris versicolor
fig = plt.figure(figsize=(7,4))
eixo = fig.add_axes([0,0,1,1])

#divide especies
df_iv = df[df['espécie'] == 'Iris-versicolor']

mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

eixo.hist(df_iv['comprimento_pétala'], bins=20)
eixo.set_title('Iris Versicolor', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
               xy=(4.6, 6), fontsize=20)

#media
eixo.axvline(mu, color = 'red', linestyle = '--')
eixo.annotate('média', xy=(mu-0.5,5.5), fontsize= 20,color='red')

#mediana
eixo.axvline(df_iv['comprimento_pétala'].median(), color = 'orange', linestyle='--')
eixo.annotate('mediana', xy=(df_iv['comprimento_pétala'].median()-0.7, 6.5), fontsize=18, 
              color='orange')

#salvar
fig.savefig('histograma_iv.png', bbox_inches='tight')
```

```python
#para setosa
fig = plt.figure(figsize=(7,4))
eixo = fig.add_axes([0,0,1,1])

#divide especies
df_is = df[df['espécie'] == 'Iris-setosa']

mu, sigma = df_is['comprimento_pétala'].mean(), df_is['comprimento_pétala'].std()

eixo.hist(df_is['comprimento_pétala'], bins=20)
eixo.set_title('Iris setosa', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
             xy = (1.7,7), fontsize = 20)

#media
eixo.axvline(mu, color = 'red', linestyle = '--')
eixo.annotate('média', xy=(mu-0.2,13.5), fontsize= 20,color='red')

#mediana
eixo.axvline(df_is['comprimento_pétala'].median(), color = 'orange', linestyle='--')
eixo.annotate('mediana', xy=(df_is['comprimento_pétala'].median()+0.05, 12), fontsize=20,
             color='orange')

#salvar
fig.savefig('histograma_is.png', bbox_inches='tight')
```

```python
#para flores virginica
fig = plt.figure(figsize=(7,4))
eixo = fig.add_axes([0,0,1,1])

#divide especies
df_ivc = df[df['espécie'] == 'Iris-setosa']

mu, sigma = df_ivc['comprimento_pétala'].mean(), df_ivc['comprimento_pétala'].std()

eixo.hist(df_ivc['comprimento_pétala'], bins=20)
eixo.set_title('Iris virginica', fontsize=15, pad=10)
eixo.set_xlabel('Comprimento da pétala', fontsize=15)
eixo.grid(True)

eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
             xy = (6.3,7), fontsize = 20)

#media
eixo.axvline(mu, color = 'red', linestyle = '--')
eixo.annotate('média', xy=(mu,8), fontsize= 20,color='red')

#mediana
eixo.axvline(df_ivc['comprimento_pétala'].median(), color = 'orange', linestyle='--')
eixo.annotate('mediana', xy=(df_is['comprimento_pétala'].median(), 7), fontsize=20,
             color='orange')

#salvar
fig.savefig('histograma_ivg.png', bbox_inches='tight')
```

```python
#uso de imagens
from PIL import Image

largura, altura = Image.open('histograma_ivg.png').size
combinada = Image.new("RGB", (3*largura, altura))
intervalo = 0
for imagem in map(Image.open, ['histograma_iv.png', 'histograma_is.png',
                               'histograma_ivg.png']):
    combinada.paste(imagem, (intervalo, 0))
    intervalo += largura

combinada.save('combinada.png')
```
